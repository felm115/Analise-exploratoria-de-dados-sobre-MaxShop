# Visualização de dados com Matplotlib e Python

Este notebook usa o arquivo tratado `MaxShop_tratado.xlsx` para gerar:
- Gráfico de barras (faturamento por ano)
- Gráfico de colunas (faturamento por estado)
- Gráfico de pizza (faturamento por região)

```python
import pandas as pd
import matplotlib.pyplot as plt

INPUT_PATH = 'data/processed/MaxShop_tratado.xlsx'

df = pd.read_excel(INPUT_PATH)

df['Data_Venda'] = pd.to_datetime(df['Data_Venda'], dayfirst=True, errors='coerce')
df['Ano'] = df['Data_Venda'].dt.year

df.head()
```

## 1. Faturamento por Ano

```python
faturamento_ano = (
    df.groupby('Ano')['Faturamento']
    .sum()
    .reset_index()
)

plt.figure(figsize=(8,5))
plt.bar(faturamento_ano['Ano'].astype(str), faturamento_ano['Faturamento'])
plt.title('Faturamento por Ano')
plt.xlabel('Ano')
plt.ylabel('Faturamento')
plt.show()
```

## 2. Faturamento por Estado

```python
faturamento_estado = (
    df.groupby('Estado')['Faturamento']
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(10,5))
plt.bar(faturamento_estado.index, faturamento_estado.values)
plt.title('Faturamento por Estado')
plt.xlabel('Estado')
plt.ylabel('Faturamento')
plt.xticks(rotation=45)
plt.show()
```

## 3. Faturamento por Região

```python
faturamento_regiao = (
    df.groupby('Regiao')['Faturamento']
    .sum()
)

plt.figure(figsize=(7,7))
plt.pie(
    faturamento_regiao.values,
    labels=faturamento_regiao.index,
    autopct='%1.1f%%'
)
plt.title('Faturamento por Região')
plt.show()
```
